```python
#importing Turtle Graphics to Python
import turtle

#Set up a function that draws a rectangle
def draw_rectangle(width, height, r,g,b, x = 0, y = 0):
    t2 = turtle.Turtle() 
    t2.speed(250)
    t2.fillcolor(r,g,b) 
    t2.hideturtle()
    t2.up()
    t2.goto(x,y)
    t2.down()
    t2.begin_fill()
    
    #directions for drawing rectangle
    for i in range(2):
        t2.forward(width) 
        t2.right(90)
        t2.forward(height)
        t2.right(90)
    t2.end_fill()
    t2.setx(width)
    
#Setting up a function that draws stars    
def draw_star(size, r,g,b, x = 0, y = 0):
    t2 = turtle.Turtle()
    t2.speed(250)
    t2.pencolor(get_color("white"))
    t2.up()
    t2.goto(x,y)
    t2.hideturtle()
    t2.down()
    t2.fillcolor(r,g,b)
    t2.begin_fill()
    
    #calculation of the angles in order to draw five-end star
    for i in range(5):
        t2.right(288)
        t2.forward(size)
        t2.right(144)
        t2.forward(size)
    t2.end_fill()

#Setting up the function that determines rgb colors   
def get_color(color):
    if color == "blue":
        return 0,0,1
    if color == "red":
        return 1,0,0
    if color == "white":
        return 1,1,1
    else:
        return 0,0,0

#creating the function that draws the flag using the functions created before
def draw_flag(height):
    flag_width = 1.9 * float(height) #flag width
    square_height = height * (7/13) #blue rectangle height
    square_width = flag_width * (2/5) #blue rectangle width 
    stripe = height / 13 #stripe height 
    star = stripe * (2/7) #star size
    v = square_height / 9 #parameter for horizontal location of stars
    w = square_width / 14 #parameter for vertical location of stars
    
    #drawing blue rectangle
    r,g,b = get_color("blue")
    draw_rectangle(square_width, square_height, r,g,b, 0, height)
    
    #drawing stripes next to blue rectangle
    r,g,b = get_color("red")
    draw_rectangle(flag_width - square_width, stripe, r,g,b, square_width, height)
    r,g,b = get_color("white")
    draw_rectangle(flag_width - square_width, stripe, r,g,b, square_width, height-stripe)
    r,g,b = get_color("red")
    draw_rectangle(flag_width - square_width, stripe, r,g,b, square_width, height-stripe*2)
    r,g,b = get_color("white")
    draw_rectangle(flag_width - square_width, stripe, r,g,b, square_width, height-stripe * 3)
    r,g,b = get_color("red")
    draw_rectangle(flag_width - square_width, stripe, r,g,b, square_width, height-stripe * 4)
    r,g,b = get_color("white")
    draw_rectangle(flag_width - square_width, stripe, r,g,b, square_width, height-stripe * 5)
    r,g,b = get_color("red")
    draw_rectangle(flag_width - square_width, stripe, r,g,b, square_width, height-stripe * 6)
    
    #drawing stripes below blue rectangle 
    r,g,b = get_color("white")    
    draw_rectangle(flag_width, stripe, r,g,b, 0, height-stripe*7)
    r,g,b = get_color("red")
    draw_rectangle(flag_width, stripe, r,g,b, 0, height-stripe * 8)
    r,g,b = get_color("white")
    draw_rectangle(flag_width, stripe, r,g,b, 0, height-stripe * 9)
    r,g,b = get_color("red")
    draw_rectangle(flag_width, stripe, r,g,b, 0, height-stripe * 10)
    r,g,b = get_color("white")
    draw_rectangle(flag_width, stripe, r,g,b, 0, height-stripe * 11)
    r,g,b = get_color("red")
    draw_rectangle(flag_width, stripe, r,g,b, 0, height-stripe * 12)

    #drawing stars
    r,g,b = get_color("white")
    #drawing the rows of 6 stars
    for i in range(1,13,2):
        for j in range(1,10,2):
            draw_star(star, r,g,b, v * i , height-(w * j) )
    #drawing the rows of 5 stars
    for i in range(2,12,2):
        for j in range(2,10,2):
            draw_star(star,r,g,b, v * i, height-(w * j))
    #finishing the drawing and exiting
    turtle.done()
    turtle.Screen().bye()
    turtle.Screen().exitonclick()
    
```

```python
 #creating main function that calls all four funtions together
def main():
    draw_flag(200)
    turtle.done()

if __name__ == "__main__":
    main()
```
