```python
f = open("numbers.txt", "r")
data = f.read()

Nums = data.replace("\n", ",").split(",")

for i in range(len(Nums)):
    Nums[i] = int(Nums[i])

numsFrequency = {}

for number in Nums:
    if number in numsFrequency:
        numsFrequency[number] += 1
    else:
        numsFrequency[number] = 1
        
print(numsFrequency)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.bar(list(numsFrequency.keys()), numsFrequency.values(), color='y')
plt.show()
```

```python
import json

json_elements = json.dumps(numsFrequency, indent = 4)
print(json_elements)

with open('json_elements.json', 'w') as outfile:
    json.dump(numsFrequency, outfile)
```

```python
import pandas as pd

df = pd.read_csv('C:/Users/micha/Downloads/Computing Midterm Project/amazon-orders.csv')
df.head((15))
```

```python
# Cleaning Data: Replacing the absence of information with 0

df = df.fillna(0) 
df.head((15))

# Change datetime data type so that the computer can recognize them

df['Order Date'] = pd.to_datetime(df['Order Date']) 
df.head((15))
```

```python
# Identifying the months in my Data

df.loc[0, 'Order Date'].month_name()
```

```python
# Making a column of months and checking

df['Month'] = df['Order Date'].dt.month_name()

df.head(15)
```

```python
# Statistical Analysis with Dates

print('Total amount spent on Amazon: $', df["Total Charged"].sum())

print('Average spending per order: $', df["Total Charged"].mean())

print('Median of my purchases: $', df["Total Charged"].median())

print('My highest-priced purchase: $', df["Total Charged"].max())

print('My lowest-priced purchase: $', df["Total Charged"].min())

print('Standard Deviation of my purchases:', df["Total Charged"].std())
```

```python
# Daily Orders

daily_orders = df.groupby('Order Date').sum()["Total Charged"]
daily_orders.head((15))
```

```python
# Bar chart of amount spent PER DAY of order

daily_orders.plot.bar(color='#61D199')
plt.title("Amount Spent PER DAY of Order")
```

```python
# Monthly Summary

import datetime
df["Month"] = pd.DatetimeIndex(df["Order Date"]).month

monthlyCost = df.groupby("Month")["Total Charged"].describe()
print(monthlyCost)
```
